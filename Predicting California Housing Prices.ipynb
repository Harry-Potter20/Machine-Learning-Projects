```python
import pandas as pd

housing_data = pd.read_csv("housing.csv")
housing_data.head()
```

```python
housing_data['ocean_proximity'].value_counts()
```

```python
#randomise the dataset in order to improve the predictive outcome
housing_data_shuffled = housing_data.sample(n=len(housing_data), random_state=1)
housing_data_shuffled
```

```python
#transform categorical data to numerical data to increase model accuracay
pd.get_dummies(housing_data_shuffled['ocean_proximity'].head())
```

```python
#drop the 'ocean_proximity' column
housing_data_shuffled.drop('ocean_proximity', axis=1).head()
```

```python
#add the dammy colums created to the housing_data dataframe
housing_df = pd.concat([housing_data_shuffled.drop('ocean_proximity', axis=1),
                        pd.get_dummies(housing_data_shuffled['ocean_proximity'])], axis=1)
housing_df
```

```python
housing_df = housing_df[['longitude', 'latitude',
                         'housing_median_age',	'total_rooms',
                         'total_bedrooms',	'population',
                         'households',	'median_income',
                         	'<1H OCEAN',
                         'INLAND',	'ISLAND',
                         'NEAR BAY',	'NEAR OCEAN', 'median_house_value']]

housing_df
```

```python
housing_df.isnull().sum()
```

```python
#fill null values in 'total_bedrooms' column with the mean values'
housing_df['total_bedrooms'] = housing_df['total_bedrooms'].fillna(housing_df['total_bedrooms'].mean())

housing_df.isnull().sum()
```

```python
#splitting the massive datasets into training, testing and validation sets
train_df, test_df, val_df = housing_df[:18000], housing_df[18000:19217], housing_df[19215:]
len(train_df), len(test_df), len(val_df)
```

```python
X_train, y_train = train_df.to_numpy()[:, :-1], train_df.to_numpy()[:, -1]
X_val,  y_val = val_df.to_numpy()[:, :-1], val_df.to_numpy()[:, -1]
X_test,  y_test = test_df.to_numpy()[:, :-1], test_df.to_numpy()[:, -1]
```

```python
y_train.shape
```

```python
#standardizing the data
from sklearn.preprocessing import StandardScaler
import numpy as np

scaler = StandardScaler().fit(X_train[:, :8])

def preprocessor(X):
  A = np.copy(X)
  A[:, :8] = scaler.transform(A[:, :8])
  return A

X_train, X_val, X_test = preprocessor(X_train), preprocessor(X_val), preprocessor(X_test)
```
